```python
from numpy.random import randn  as rd
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=rd(500).cumsum()))
df.head()
```

```python
sns.relplot(x="time",y="value",kind="line",data=df)
```

# line plot, with sorted data

```python
sns.relplot(x="time",y="value",kind="line",data=df,sort=True)
```
